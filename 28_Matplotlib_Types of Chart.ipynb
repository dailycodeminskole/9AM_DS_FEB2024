```python
import numpy as np
import matplotlib.pyplot as plt
```

# BAR

```python
# shop1
# fruit1 : 10
# fruit2 : 20
# fruit3 : 5
```

```python
plt.bar([1,2,3],[10,20,5])
plt.show()
```

```python
plt.bar([1,2,3],[10,20,5],width=0.5) #reducing the default thickness
plt.legend("Shop1") # adding legend
plt.show()
```

```python
# shop2
# fruit1 : 5
# fruit2 : 10
# fruit3 : 6
```

```python
plt.bar([1,2,3],[5,10,6])
plt.legend("Shop2")
plt.show()
```

```python
# plotting both the charts in single canvas
```

```python
plt.bar([1,2,3],[10,20,5]) #reducing the default thicknes
plt.bar([1,2,3],[5,10,6],width=0.5)
plt.legend(["Shop1",'Shop2'])
plt.show()
```

```python
plt.bar([1,2,3],[10,20,5],width=0.5) #reducing the default thicknes
plt.bar([1.5,2.5,3.5],[5,10,6],width=0.5)
plt.legend(["Shop1",'Shop2'])
plt.show()
```

```python
plt.bar([1,2,3],[10,20,5],width=0.3) #reducing the default thicknes
plt.bar([1.3,2.3,3.3],[5,10,6],width=0.3)
plt.legend(["Shop1",'Shop2'])
plt.show()
```
