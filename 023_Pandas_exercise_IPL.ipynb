# Freq counts over a categorical data

```python
# data :  Numerical : large variation 
#         Categorical : less variation
```

```python
data = pd.read_csv('matches.csv')
```

```python
data['city'].value_counts()
```

```python
data['team1'].value_counts()
```

```python
data['team2'].value_counts()
```

# findin the IPL winner of every season

```python
data.drop_duplicates('season',keep='last').sort_values('season')
```

```python
data.drop_duplicates('season',keep='last').sort_values('season')[['season','winner','team1','team2']]
```

```python
data.drop_duplicates('season',keep='last').sort_values('season')[['season','winner','team1','team2']].sort_values('season',ascending=False)
```
