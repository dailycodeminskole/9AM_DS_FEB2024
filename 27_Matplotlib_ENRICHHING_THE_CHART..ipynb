```python
import matplotlib.pyplot as plt
import numpy as np
```

# REVISION

```python
# using numpy to create  a cube function graph
```

```python
x_val  = np.arange(-10,10,.5)
y_val  = x_val**3
plt.xticks(np.arange(-10,10,1),rotation=90)
plt.plot(x_val,y_val)
plt.show()
```

```python
x_val = np.arange(-4,4,.1) #elementwise operation possible
y_val = x_val**2
plt.xticks(np.arange(-4,4,.3),rotation=90)
plt.plot(x_val,y_val)
plt.show()
```

#  ENRICHING THE GRAPH

```python
# changing the color
# changing the default from line to *
# 
```

```python
x_val = np.arange(-4,4,.1) #elementwise operation possible
y_val = x_val**2
plt.xticks(np.arange(-4,4,.3),rotation=90)
plt.plot(x_val,y_val,color='green')
plt.show()
```

```python
x_val = np.arange(-4,4,.1) #elementwise operation possible
y_val = x_val**2
plt.xticks(np.arange(-4,4,.3),rotation=90)
plt.plot(x_val,y_val,color='green',marker='*')
plt.show()
```

```python
x_val = np.arange(-4,4,.1) #elementwise operation possible
y_val = x_val**2
plt.xticks(np.arange(-4,4,.3),rotation=90)
plt.plot(x_val,y_val,'b*') #blue star
plt.show()
```

```python
x_val = np.arange(-4,4,.1) #elementwise operation possible
y_val = x_val**2
plt.xticks(np.arange(-4,4,.3),rotation=90)
plt.plot(x_val,y_val,'r+') #blue star
plt.show()
```

# multiple plots in the same canvas

```python
x_val = np.arange(-4,4,.1) #elementwise operation possible
plt.xticks(np.arange(-4,4,.3),rotation=90)


y1_val = x_val**2
plt.plot(x_val,y1_val)

y2_val  = x_val**3
plt.plot(x_val,y2_val)


plt.show()
```

# adding legends

```python
x_val = np.arange(-4,4,.1) #elementwise operation possible
plt.xticks(np.arange(-4,4,.3),rotation=90)


y1_val = x_val**2
y2_val  = x_val**3


plt.plot(x_val,y1_val,label="Square")
plt.plot(x_val,y2_val,label="Cube")

plt.legend()
plt.show()
```

```python
x_val = np.arange(-4,4,.1) #elementwise operation possible
plt.xticks(np.arange(-4,4,.3),rotation=90)


y1_val = x_val**2
y2_val  = x_val**3


plt.plot(x_val,y1_val,)
plt.plot(x_val,y2_val,)

plt.legend(['Square','Cube'],loc='lower right') #chnaging the default to lower right
plt.show()
```

```python
x_val = np.arange(-4,4,.1) #elementwise operation possible
plt.xticks(np.arange(-4,4,.3),rotation=90)


y1_val = x_val**2
y2_val  = x_val**3


plt.plot(x_val,y1_val,)
plt.plot(x_val,y2_val,)

plt.legend(['Square','Cube'],loc='lower right',frameon=False) #chnaging the default to lower right
plt.show()
```

```python
x_val = np.arange(-4,4,.1) #elementwise operation possible
plt.xticks(np.arange(-4,4,.3),rotation=90)


y1_val = x_val**2
y2_val  = x_val**3


plt.plot(x_val,y1_val,)
plt.plot(x_val,y2_val,)

plt.legend(['Square','Cube'],loc='lower right',frameon=False,ncol=2) #chnaging the default to lower right
plt.show()
```

```python
x_val = np.arange(-4,4,.1) #elementwise operation possible
plt.xticks(np.arange(-4,4,.3),rotation=90)


y1_val = x_val**2
y2_val  = x_val**3


plt.plot(x_val,y1_val,)
plt.plot(x_val,y2_val,)

plt.legend(['Square','Cube'],loc='lower right',frameon=False,ncol=2) #chnaging the default to lower right

plt.text(0,0+.5,'(0,0)')
plt.text(1,1+.5,'two')


plt.xlim(-2,2)
plt.ylim(-2,2)

plt.show()
```
