# subplot using Objects

```python
x = np.arange(1,10,.1)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

plt.figure()
plt.subplot(3,1,1)
plt.plot(x,y1)

plt.subplot(3,1,2)
plt.plot(x,y2)


plt.subplot(3,1,3)
plt.plot(x,y3)
```

```python
x = np.arange(1,10,.1)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

plt.figure()
plt.subplot(1,3,1)
plt.plot(x,y1)

plt.subplot(1,3,2)
plt.plot(x,y2)


plt.subplot(1,3,3)
plt.plot(x,y3)
```

```python
x = np.arange(1,10,.1)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

plt.figure()
plt.subplot(2,3,1)
plt.plot(x,y1)

plt.subplot(2,3,2)
plt.plot(x,y2)


plt.subplot(2,3,3)
plt.plot(x,y3)


plt.subplot(2,3,4)
plt.plot()

plt.subplot(2,3,5)
plt.plot()

plt.subplot(2,3,6)
plt.plot()
```

# 3d plotting

```python
a = np.arange(-1,1,0.005)
b = a

a ,b  = np.meshgrid(a,b)
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(a,b,a**2+b**2)
```
